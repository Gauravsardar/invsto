```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
data = yf.download("AAPL", start="2023-01-01", end="2023-02-01")
```

```python
# Printing the first 5 entries of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Handle missing values (e.g., remove rows with missing values)
data.dropna(inplace=True)
```

```python
# Outlier removal
from scipy import stats

z_scores = stats.zscore(data)
threshold = 3
data = data[(z_scores < threshold).all(axis=1)]
```

```python
# Check data types of each column
data_types = data.dtypes

# Filter out rows containing string values in any column
data = data[(data.applymap(lambda x: type(x) != str)).all(axis=1)]
```

```python
# Converting to pandas dataframe
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame containing OHLC data with a datetime index

# User input for the number of days for moving averages calculation
num_days = int(input("Enter the number of days for moving averages calculation: "))

# Calculate Simple Moving Average (SMA)
df['SMA'] = df['Close'].rolling(window=num_days, min_periods=1).mean()

# Calculate Exponential Moving Average (EMA)
df['EMA'] = df['Close'].ewm(span=num_days, adjust=False).mean()

# Calculate Weighted Moving Average (WMA)
weights = np.arange(1, num_days + 1)

def wma_func(prices, weights):
    prices = np.asarray(prices)
    weights = np.asarray(weights)[:len(prices)]
    return np.dot(prices, weights) / weights.sum()

df['WMA'] = df['Close'].rolling(window=num_days, min_periods=1).apply(wma_func, args=(weights,), raw=True)

# Display the DataFrame with the calculated indicators
df.head()
```
